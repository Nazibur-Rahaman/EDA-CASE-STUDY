# 2. Missing values and Best Metric for Data Imputaion

Here Best Metric to Impute the Missing Values

'NAME_TYPE_SUITE'                  
'OBS_30_CNT_SOCIAL_CIRCLE'         
'DEF_30_CNT_SOCIAL_CIRCLE'         
'OBS_60_CNT_SOCIAL_CIRCLE'         
'DEF_60_CNT_SOCIAL_CIRCLE'

'AMT_GOODS_PRICE'                  
'AMT_ANNUITY'                      
'CNT_FAM_MEMBERS'                  
'DAYS_LAST_PHONE_CHANGE'           

This columns have few missing values percentage. so can try to impute the missing values with best metric.

```python
#this is categorical columns , we will impute this with mode values.

values=newapp_data['NAME_TYPE_SUITE'].mode()

newapp_data["NAME_TYPE_SUITE"].fillna(values,inplace=True)
```

```python
#outliers is present so we can impute with median values,as in presence of outliers median values will not effect much

values=newapp_data['AMT_ANNUITY'].median()

newapp_data["AMT_ANNUITY"].fillna(values,inplace=True)
```

```python

values=newapp_data['OBS_30_CNT_SOCIAL_CIRCLE'].median()

newapp_data['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(values,inplace=True)
```

```python
#Again the same ouliers reason we can use median , as asset value can effect with huge outliers

values=newapp_data['AMT_GOODS_PRICE'].median()

newapp_data["AMT_GOODS_PRICE"].fillna(values,inplace=True)
```

```python
# family members can easily effect with ouliers so can impute with median
values=newapp_data['CNT_FAM_MEMBERS'].median()

newapp_data['CNT_FAM_MEMBERS'].fillna(values,inplace=True)
```

```python
# accompanying during loan application , categorical values, single , with family, so we use mode values.

values=newapp_data['NAME_TYPE_SUITE'].mode()

newapp_data['NAME_TYPE_SUITE'].fillna(values,inplace=True)
```

```python
# external source2 which is from bank itself, so imputed with median values.

values=newapp_data['EXT_SOURCE_2'].median()

newapp_data['EXT_SOURCE_2'].fillna(values,inplace=True)
```

```python
# Gender distribution have 4 missing values
newapp_data['CODE_GENDER'].value_counts()
```

```python
# can impute gender column with Female, as it has female applicant more
newapp_data.loc[newapp_data['CODE_GENDER'] == 'XNA','CODE_GENDER'] = 'F'
newapp_data['CODE_GENDER'].value_counts()
```

```python
#oocupation column have 96391 missing values
newapp_data['OCCUPATION_TYPE'].value_counts(dropna=False)
```

```python
# make this missing occupation a new column, becuse for huge missing values, so it will not biased.
newapp_data["OCCUPATION_TYPE"].fillna("Missing Occ. Category",inplace=True)
```
